```python
def gcdOfStrings(str1, str2):
    if str1 == str2: return str1
    gcd = ""
    i = 0
    if len(str1)<len(str2):
        str11,str22 = str2,str1
    else:
        str11,str22 = str1,str2
    while i<len(str11) :
        if str1[i] in str22:
            j = str22.index(str11[i])
            gcd+=str22[j]
            maxlen = len(str22)-j
            for l in range(1,maxlen):
                if str11[i+l]==str22[j+l]:
                    gcd+=str22[j+l]
                else:
                    gcd=""
                    break
            if len(gcd)==1: 
                gcd=""
            return gcd
        i+=1
        
    return gcd
```

```python
def powerfulIntegers(x, y, bound):
#     if bound >= 2: out.append(2)
#     if bound >= x: out.append(x)
#     if bound >= y: out.append(y)

    out = []
    
    if x==1 and y==1: return [2]
    if x==1: x,y=y,x
    
    i=0
    while i>=0:
        j=0
        re = pow(x,i)+pow(y,j)
        while re<=bound:
            out.append(re)
            j+=1
            if y==1: break
        i+=1
        if pow(x,i)>bound:
            return list(set(out))  
```

```python
def isPowerOfThree(n):
    """
    :type n: int
    :rtype: bool
    """
    if n==0 or n==2: return False
    if n==1: return True
    while n%3==0:
        n=int(n/3)
    if n==1: return True
    else: return False
```

```python
def isPowerOfTwo(n):
    return n!=0 and (n&(n-1)==0)
```
